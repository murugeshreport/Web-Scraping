```python
import requests
from bs4 import BeautifulSoup
```

```python
for page in range(1,2):
    url = "https://www.gopgo.in/bengaluru/pgo-properties?city_id=689&latitude=12.971599&longitude=77.594563&search_by=city_id&page={0}&total_properties=3763".format(page)
    respanse = requests.get(url = url).content
```

```python
soup = BeautifulSoup(respanse)
```

```python
hostal_list = (soup.find_all('div',{'class':'proptery-hostels-list'}))
```

```python
data = []
```

```python
for i in hostal_list:
    name = i.find('h2',{'class':'proptery-hostels-name-l-title'})
    name = name.text if name else ''
    name = name.strip()
    
    gender = i.find('span',{'class':'genderTag boys'})
    gender = gender.text if gender else ''
    gender = gender.strip()
    
    add = i.find('div',{'class':'proptery-hostels-p'})
    add = add.text if add else ''
    add = add.strip()
    
    price = i.find('p',{'class':'property-card-col__monthly'})
    price = price.text if price else ''
    price = price.strip()
    
    dayp = i.find('p',{'class':'property-card-col__daily'})
    dayp = dayp.text if dayp else ''
    dayp = dayp.strip()
   
    data.append([name,gender,add,price,dayp])
    
```

```python
data
```

```python
data = pd.DataFrame(data)
```
