```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
import requests,openpyxl
```

# first page

```python
jobs = []
```

```python
response = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Science&txtLocation=')
soup = BeautifulSoup(response.text,'html.parser')
   
```

```python
value = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
```

```python
for i in value:
    name = i.find('h3',class_='joblist-comp-name').text
    skill = i.find('span',class_='srp-skills').text

    
    jobs.append([name,skill])
```

```python
m = pd.DataFrame(jobs)
```

```python
m.columns=['company','skill']
```

```python
m
```

```python
def clean_html(html):

    # parse html content
    soup = BeautifulSoup(html, "html.parser")

    for data in soup(['style', 'script', 'code', 'a']):
        # Remove tags
        data.decompose()

    # return data by retrieving the tag content
    return ' '.join(soup.stripped_strings)
```

```python
# To remove HTML first and apply it directly to the source text column.
m['company'] = m['company'].apply(lambda x: clean_html(x))
```

```python
# To remove HTML first and apply it directly to the source text column.
m['skill'] = m['skill'].apply(lambda x: clean_html(x))
```

# multiple page

```python
pages = np.arange(1,100)
```

```python
data1 = []
```

```python
for page in pages:
    res =  requests.get("https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=Data%20Science&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=data%20science&pDate=I&sequence="+str(page)+"&startPage=1")
    soup = BeautifulSoup(res.text,'html.parser')
    values = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
    
    for jobs in values:
        name = jobs.find('h3',class_='joblist-comp-name').text
        skill = jobs.find('span',class_='srp-skills').text
        data1.append([name,skill])
    
    
    
```

```python
data1 = pd.DataFrame(data1)
```

```python
data1.columns=['company','skills']
```

```python
# To remove HTML first and apply it directly to the source text column.
data1['company'] = data1['company'].apply(lambda x: clean_html(x))
```

```python
# To remove HTML first and apply it directly to the source text column.
data1['skills'] = data1['skills'].apply(lambda x: clean_html(x))
```

```python
data1.to_csv('Jobs_skills_data_set.csv',index = False)
```
