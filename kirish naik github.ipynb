```python
data =[]
```

```python
for page in pages:
    res =  requests.get("https://github.com/krishnaik06?page="+str(page)+"&tab=repositories")
    soup = BeautifulSoup(res.text,'html.parser')
    m = soup.find_all('li',class_="col-12 d-flex flex-justify-between width-full py-4 border-bottom color-border-muted public source")
    
    
    for i in m:
        title =i.find('h3',class_="wb-break-all").a.text
        star = i.find('div',class_='f6 color-fg-muted mt-2').a.text
        up = i.find('relative-time',class_='no-wrap').text

        data.append([title,star,up])
    
   
    
    
    
```

```python
data
```

```python
def clean_html(html):

    # parse html content
    soup = BeautifulSoup(html, "html.parser")

    for data in soup(['style', 'script', 'code', 'a']):
        # Remove tags
        data.decompose()

    # return data by retrieving the tag content
    return ' '.join(soup.stripped_strings)
```

```python
data = pd.DataFrame(data)
```

```python
data.columns=['title','stars','updated']
```

```python
data
```

```python
data['title'] = data['title'].apply(lambda x: clean_html(x))
```

```python
data['stars'] = data['stars'].apply(lambda x: clean_html(x))
```

```python
data['updated'] = data['updated'].apply(lambda x: clean_html(x))
```
